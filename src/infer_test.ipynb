```python
import cv2
from ultralytics import YOLO
import sys

model = YOLO('/workspace/yolov8/dev/results/x_model_epoch20_batch16/weights/best.pt')

video_path = "/workspace/yolov8/dev/test/video/test1.avi"
output_path = "/workspace/yolov8/dev/test/video/"
cap = cv2.VideoCapture(video_path)

fps = 15
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fcc = cv2.VideoWriter_fourcc(*'mp4v')

out = cv2.VideoWriter(output_path+"output.mp4", fcc, fps, (width, height))

if not out.isOpened():
    print("File open failed!")
    cap.release()
    sys.exit(1)

while cap.isOpened():
    ret, frame = cap.read()
    
    if ret:
        result = model.predict(frame, device=0, conf=0.7)
        
        annotated_frame = result[0].plot()
        
        out.write(annotated_frame)
        
    else:
        break
    
cap.release()
out.release()
cv2.destroyAllWindows()
```
